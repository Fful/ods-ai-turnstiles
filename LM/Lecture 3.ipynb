```python
import pandas as pd
import numpy as np
```

```python
def make_harmonic_features(value, period=24): 
    value *= 2 * np.pi / period 
    return np.cos(value), np.sin(value)
```

```python
time_list = []

for time in range(0,25):
    print(time, make_harmonic_features(time))
    time_list.append(make_harmonic_features(time))
```

```python
import matplotlib.pyplot as plt
```

```python
a = plt.figure()
plt.plot(range(0,25), time_list)
plt.xlabel('time')
```
